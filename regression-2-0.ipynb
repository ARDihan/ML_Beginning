**Linear Regression**

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats  #a function from scipy library to use in linear algbr

x=np.random.randint(1,20,size=10)
y=np.random.randint(1,20,size=10)

slope,intercept,r,p,std_err=stats.linregress(x,y) #dont know, may be it does lin regression

def fun1(x):
    return slope * x + intercept #from the y=mx+c
fun1(x)
```

```python
model_1=list(map(fun1,x))
plt.scatter(x,y)
plt.plot(model_1,x)
```

```python
import pandas as pd

df = pd.read_csv("/kaggle/input/breast-cancer-dataset/Breast_cancer_dataset.csv")


import warnings
warnings.filterwarnings("ignore", category=RuntimeWarning)

df.head(10)
```

```python
x=df[['area_mean','radius_mean']]

y=df['texture_worst']
```

```python
from sklearn import linear_model
regr=linear_model.LinearRegression()
regr.fit(x,y)
```

```python
predicted_texture_worst=regr.predict([[1000,20]])
print(predicted_texture_worst)

print("regression coefficient:",regr.coeff_) # predict + (coefff X x)
```
