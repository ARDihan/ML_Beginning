**Multiple regression****

```python
import matplotlib.pyplot as plt
import numpy as np

num1=np.random.randint(1,30,size=10)
num2=np.random.randint(1,30,size=10)

x=np.sort(num1) # sorting the random variables and storing in x & y
y=np.sort(num2)

plt.scatter(x,y)
```

```python
import matplotlib.pyplot as plt
import numpy as np

my_model=np.poly1d(np.polyfit(x,y,4))
my_line=np.linspace(1,22,200)

plt.scatter(x,y)
plt.plot(my_line,my_model(my_line))
plt.show()
```

```python
from sklearn.metrics import r2_score
print(r2_score(y,my_model(x)))

'''From sklearn library we will use:
        1. sklearn.model_selection.train_test_split
        2. sklearn.linear_model.LinearRegression
        3.Errors:
            sklearn.metrics.mean_absolute_error
            sklearn.metrics.mean_squared_error
            sklearn.metrics.r2_score'''
```

Full Code of Regression with Algorithms

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error , mean_squared_error

# 1.Read the file 
df=pd.read_excel('/kaggle/input/heart-disease-dataset-3k-rows-python-code-2025/heart_disease_dataset.xlsx')

print(df)
```

```python
# 2. define the input and output features

x= df[['age','chol','age','oldpeak','cp'  ,'trestbps']]
y= df['heart_disease']

# 3. import the train function and define the test and train percentage

x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=40)

# 4.Create the model and train it by fitting the model using LinearRegression function(OLS system)

my_model= LinearRegression() #create the model and name it like my_model
my_model.fit(x_train,y_train)

# 5. Inspect Parameters

coefficient= my_model.coef_
interception=my_model.intercept_

print("Coefficient is: ",coefficient)
print("Intercept = ",interception)

# 6. Predict the data, see the output of the trained model

y_pred = my_model.predict(x_test)

mae =mean_absolute_error(y_test,y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)
print("R Score:", r2)

import joblib
joblib.dump(my_model,"My_First_Model.pkl")
```
